Статья:  
[AA] Learning Data Augmentation Strategies for Object Detection | [arXiv' 19]  
https://arxiv.org/pdf/1906.11172.pdf

В этой статье инженеры описывают, три способа, которые улучшают обучение нейронной сети на датасете COCO.
• Цветовые операции. Искажаются цветовые каналы, не влияя на расположение ограничивающих рамок (например, Equalize, Contrast, Brightness).  
• Геометрические операции. Геометрическое искажение изображения, которое соответственно изменяет местоположение и размер аннотаций ограничивающего прямоугольника (например, Rotate, ShearX, TranslationY и т. Д.).  
• Операции с ограничивающими рамками. Искажается только содержимое пикселей, содержащееся в аннотациях ограничительной рамки (например, BBox Only Equalize, BBox Only Rotate, BBox Only FlipLR).  
Для любых операций, которые влияют на геометрию изображения, они также изменили размер и расположение ограничивающего прямоугольника, чтобы сохранить согласованность.
