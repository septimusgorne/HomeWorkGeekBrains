```python
df = pd.read_csv(r'kc_house_data.csv', encoding='cp1251')
```

<b>1.1 Создать новый признак price_per_sq_lot, который будет содержать стоимость за один кв. метр общей площади</b>

```python
df['price_per_sq_lot'] = df['price']/df['sqft_lot']
df.head()
```

<b>1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0</b>

```python
df['delta_renovated'] = df['yr_renovated']/df['yr_built']
df.head()
```

```python
df['delta_renovated'] = df['delta_renovated'].apply(lambda x: x if x > 0 else 0)
df.head()
```

<b>1.3 Создайте признаки года продажи, месяца продажи</b>

```python
df['year']= df['date'].apply(lambda x: int(x[4]))
df['month']= df['date'].apply(lambda x: int(x[4:6]))
df.head()
```

<b>1.4 Удалите признаки date, zipcode, lat, long</b>

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```
