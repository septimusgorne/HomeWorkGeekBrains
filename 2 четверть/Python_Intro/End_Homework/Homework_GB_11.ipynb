```python
from sympy import *
from sympy.plotting import plot
import matplotlib
init_printing()
plot.set_linestyle ='dashed'
#plot.style.use('seaborn')
```

```python
x = Symbol('x')
```

```python
#y = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2+10*x-30
y = -5*x**2 + 10*x + 150
```

```python
#выведение графика
plot(y,(x, -10, 10), show = True)
#plot(y)
print(y)
```

```python
plot(diff(y))
```

```python
#определение корней
#производная
diff(y)
```

```python
# Нахождение корней
solve(y)
```

```python
# #нахождение вершины
# Производная
solveset(diff(y))
```

```python
#Промежуток, где f > 0
solve(y > 0)
```

```python
#Промежуток, где f < 0
solve(y < 0)
```

```python
#Интернвал на котором функция возрастает
solve(diff(y)>0)
```

```python
#Интернвал на котором функция убывает
solve(diff(y)<0)
```

```python
#Вывод интервала убывания диапазона x(-1, 20)
plot(y, (x, 20, -1))
```

```python
# Вывод интервала возрастания диапазона x(-20, 1)
plot(y, (x, 1, -20))
```
